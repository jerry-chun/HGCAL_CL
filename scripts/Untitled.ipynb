```python

event_idx = 0
energy_threshold = 0

tree = data['HGCALTBout;1']
x_t = tree["hit_x"].array()
y_t = tree["hit_y"].array()
z_t = tree["hit_z"].array()
e_t = tree["hit_Edep"].array()
layer_id = tree['hit_layer'].array()
shower_id = tree["hit_showerid"].array()
pdg = tree['PrimaryPDGIDs'].array()

x_all = np.array(x_t[event_idx])
y_all = np.array(y_t[event_idx])
z_all = np.array(z_t[event_idx])
e_all = np.array(e_t[event_idx])
layer_all = np.array(layer_id[event_idx])
sids_all = np.array(shower_id[event_idx])
pdg_all = np.array(pdg[event_idx])
print(pdg_all)

mask = e_all > energy_threshold
x = x_all[mask]
y = y_all[mask]
z = z_all[mask]
e = e_all[mask]
layers = layer_all[mask]
sids = sids_all[mask]
print(len(x))

unique_sids = np.unique(sids)
sid_to_idx = {sid: idx for idx, sid in enumerate(unique_sids)}
color_indices = np.array([sid_to_idx[sid] for sid in sids])

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(x, y, z, c=color_indices, cmap='tab20', marker='o')

# Axis labels
ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
ax.set_zlabel('z (cm)')

for sid, idx in sid_to_idx.items():
    ax.scatter([], [], [], c=[plt.cm.tab20(idx)], label=f'Shower {sid}')
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1))

plt.title(f"3D Shower Hits for Event {event_idx}\nEdep > {energy_threshold}")
plt.tight_layout()
plt.show()

```

```python

clusterer = HGCALEventClusterer_2D(dc=0.13, rhoc=0.015, dm=0.26, pPB=10, backend="cpu serial")
seps = []
puritys = []
for i in range(len(x_t)):
    x = np.array(x_t[i])
    y = np.array(y_t[i])
    z = np.array(z_t[i])
    e = np.array(e_t[i])
    layers = np.array(layer_id[i])
    sids = np.array(shower_id[i])
    
    clusterer.read_event(x, y, z, e, layers, sids=sids)
    clusterer.cluster_event(verbose=False)
    
    sep = clusterer.get_shower_barycenter_separation(weight_by_energy=False)
    pur = clusterer.return_avg_purity()
    
    seps.append(sep)
    puritys.append(pur)
 
```

```python
plt.figure(figsize=(8, 6))

# Use only 2 colors, alternating for even/odd layers
colors = (layers % 2)  
cmap = plt.get_cmap("coolwarm", 2)  

sc = plt.scatter(x, z, c=colors, cmap=cmap, s=8)

plt.xlabel("x (cm)")
plt.ylabel("z (cm)")
plt.title(f"2D x–z Hits with Alternating Layer Colors (Event {event_idx})\nEdep > {energy_threshold}")

# Colorbar shows even/odd mapping
cbar = plt.colorbar(sc, ticks=[0, 1])
cbar.ax.set_yticklabels(["Even layers", "Odd layers"])
cbar.set_label("Layer parity")

plt.tight_layout()
plt.show()
```
