```python
loader = Data_Loader(
    root="/vols/cms/mm1221/geant4sim/simulations/build/Datasets/Test_EM_2_10_delta5/Electron.10005_10006.root",
    split="test",
    max_events=100,
    batch_size=1,
    shuffle=False,
    follow_batch=["x"],
)
```

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_config = {
    "task": "contrastive",
    "hidden_dim": 64,
    "num_layers": 3,
    "dropout": 0.01,
    "k": 24,

    "contrastive_dim": 16,
    "coord_dim": 16,
    "path" : "/vols/cms/mm1221/geant4sim/scripts/training/Contrastive/runs/EM_2_10_CD16_Supcon_delta5_t01/best_model.pt"
}

model = model_loader(
    config=model_config,
    device = device
)

print(f"Model loaded successfully")
```
